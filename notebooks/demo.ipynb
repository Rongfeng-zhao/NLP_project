## Pre load the embedding model

```python
import os
import sys

os.chdir("/home/chris/Desktop/unistaff/NLP_project")
sys.path.append(os.getcwd())

from src.retriever import ProductRetriever

print(os.getcwd())

retriever = ProductRetriever(index_dir="artifacts/vector_index")
```

## query

```python
test_queries = [
    "waterproof eyebrow products with natural finish",
    "beauty products with powder item form",
    "products from Cherioll brand",
    "long lasting makeup products",
    "high rating all beauty products"
]

for query in test_queries:
    print("\nQUERY:", query)
    results = retriever.retrieve(query=query, top_k=3)

    for item in results:
        print("Rank:", item["rank"])
        print("Score:", round(item["score"], 4))
        print("Text:", item["text"][:500])
        print("-" * 80)
```
